### I wanted to see the function of `seed` and how the rb sequences of different length looks like by drawing the corresponding length of circuit and transpliling it in to the cliffords.

```python
import numpy as np

# Importing standard Qiskit libraries
from qiskit import QuantumCircuit, transpile, Aer, IBMQ
from qiskit.tools.jupyter import *
from qiskit.visualization import *
from ibm_quantum_widgets import *
from qiskit.providers.aer import QasmSimulator

# Loading your IBM Quantum account(s)
provider = IBMQ.load_account()
```

```python
import numpy as np
from qiskit_experiments.library import StandardRB, InterleavedRB
from qiskit_experiments.framework import ParallelExperiment
from qiskit_experiments.library.randomized_benchmarking import RBUtils
import qiskit.circuit.library as circuits
```

```python
provider = IBMQ.get_provider(hub='ibm-q-skku', group='skku', project='squirl')
backend = provider.get_backend('ibmq_casablanca')
```

```python
lengths = np.arange(1, 10, 1)
print(lengths)
```

```python
exp = StandardRB(qubits=[0], lengths=lengths, num_samples=1, seed=1)
```

```python
c = exp.circuits()[0]
c.draw(output='mpl')
```

```python
from qiskit import transpile
basis_gates = backend.configuration().basis_gates
transpile(c, basis_gates=basis_gates).draw(output='mpl')
```

```python
c = exp.circuits()[0]
c.draw(output='mpl')
```

```python
from qiskit import transpile
basis_gates = backend.configuration().basis_gates
transpile(c, basis_gates=basis_gates).draw(output='mpl')
```

```python
c = exp.circuits()[3]
c.draw(output='mpl')
```
